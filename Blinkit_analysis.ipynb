### **Charts Requirements**

#### **Total Sales by Fat content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.figure(figsize=(8, 6))
plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct="%1.1f%%", startangle=90, colors=colors)
plt.title("Sales by Fat Content")
plt.axis("equal")  # Equal aspect ratio for a proper circle
plt.legend(sales_by_fat.index, title="Total Sales by Fat Content", loc="best")
plt.show()
```

#### **Total Sales by Item type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)

plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_type.index, sales_by_type.values)
plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item')

for bar in bars: plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), 
                          f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8) #data labels 

plt.tight_layout()
plt.show()
```

#### **Sales Contribution by Outlet Identifier** 

```python
# Group by Outlet Identifier and calculate total sales
sales_by_outlet = ( df.groupby("Outlet Identifier")["Sales"].sum().sort_values(ascending=False).reset_index())

# Calculate cumulative sales and cumulative percentage
sales_by_outlet["Cumulative Sales"] = sales_by_outlet["Sales"].cumsum()
sales_by_outlet["Cumulative %"] = 100 * sales_by_outlet["Cumulative Sales"] / sales_by_outlet["Sales"].sum()

# Plot Pareto chart
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar plot for Sales per Outlet
bars = ax1.bar(x=range(len(sales_by_outlet)), height=sales_by_outlet["Sales"], color="skyblue")

ax1.set_xlabel("Outlet Identifier")
ax1.set_ylabel("Sales", color="skyblue")
ax1.tick_params(axis="y", labelcolor="skyblue")
ax1.set_xticks(range(len(sales_by_outlet)))
ax1.set_xticklabels(sales_by_outlet["Outlet Identifier"], rotation=45, ha="right")

# Line plot for Cumulative % of tatal sales
ax2 = ax1.twinx()
ax2.plot(range(len(sales_by_outlet)), sales_by_outlet["Cumulative %"], color="orange", marker="o", linewidth=2)
ax2.set_ylabel("Cumulative % of Total Sales", color="orange")
ax2.tick_params(axis="y", labelcolor="orange")
ax2.axhline(80, color='red', linestyle='--', linewidth=1)
ax2.text(len(sales_by_outlet) * 0.7, 82, "80% Threshold", color='red')

plt.title("Pareto Chart: Sales Contribution by Outlet Identifier")
plt.tight_layout()
plt.show()
```

#### **Sales Trend Over Time by Outlet Establishment Year** 

```python
# Convert Establishment Year to string to treat it as category for plotting
df.loc[:, "Outlet Establishment Year"] = df["Outlet Establishment Year"].astype(str)

# Group: Total Sales by Year and Outlet Type
sales_trend = df.groupby(["Outlet Establishment Year", "Outlet Type"])["Sales"].sum().unstack().fillna(0)

# Sort index so years appear in order
sales_trend = sales_trend.sort_index()

# Plot Area Chart
plt.figure(figsize=(8, 5))
sales_trend.plot(kind="area", stacked=True, figsize=(12, 6), colormap="viridis", alpha=0.8)

plt.title("Sales Trend by Outlet Establishment Year (Grouped by Outlet Type)", fontsize=14)
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales")
plt.legend(title="Outlet Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

#### **Item Type Popularity Across Outlet Types** 

```python
# Group by Item Type and Outlet Type to get total sales
pivot_data = (df.groupby(["Item Type", "Outlet Type"])["Sales"].sum().reset_index().pivot(index="Item Type", columns="Outlet Type", values="Sales").fillna(0))

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_data, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=.5)

plt.title("Item Type Popularity Across Outlet Types (by Total Sales)", fontsize=14)
plt.xlabel("Outlet Type")
plt.ylabel("Item Type")
plt.tight_layout()
plt.show()
```

```python
# Compute thresholds (quartiles)
low_sales_threshold = df["Sales"].quantile(0.25)
high_visibility_threshold = df["Item Visibility"].quantile(0.75)
low_rating_threshold = df["Rating"].quantile(0.25)

# Filter underperforming items
underperforming = df[
    (df["Sales"] <= low_sales_threshold) &
    (df["Item Visibility"] >= high_visibility_threshold) &
    (df["Rating"] <= low_rating_threshold)
]

# Optional: Show key columns only
underperforming_items = underperforming[[
    "Item Identifier", "Sales", "Item Visibility", "Rating", "Outlet Identifier", "Item Type"
]]

# Display the table
print("Underperforming Items Based on Sales, Visibility, and Rating:")
print(underperforming_items.sort_values(by="Sales").head(10))
```
